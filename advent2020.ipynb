```python
# Day 7
import re
data_day7 = re.split('\n',open("inputs/day7.txt", "r").read().strip())


def parse_bag_names(s):
    x = re.findall('([a-z]+ [a-z]+) bag|bags',s)
    y = re.findall('(\d)',s)
    y=['0'] if len(y) == 0 else y
    return x,y

def bag_dict(data):
    d = {}
    for rule in data:
        (x,y) = parse_bag_names(rule)
        l = []
        for i, color in enumerate(x[1:], start=0):
            l += [(color,int(y[i]))]
        d[x[0]]=l
    return d

all_rules = bag_dict(data_day7)
all_rules
                  
def has_color(this_bag,target_color):
    #print(this_bag)
    if this_bag == target_color:
        return True
    output = False
    for color in all_rules[this_bag]:
        if color[0] == 'no other':
            continue
        if has_color(color[0],target_color):
            output = output or True
    return output

the_bag = 'shiny gold'
count = 0
for rule in all_rules.keys():
    if has_color(rule,the_bag):
        count += 1
print(f'Part 1: {count-1} bags can contain a {the_bag} bag')

# Part 2
def get_bags_within(color):
    if color == 'no other':
        return 0
    total = 0
    for bag in all_rules[color]:
        total += bag[1]*(1+get_bags_within(bag[0]))
    return total
    
print(f'Part 2: {get_bags_within(the_bag)} bags are in a {the_bag} bag')
```

```python
# Day 8
import re
data_day8 = re.split('\n',open("inputs/day8.txt", "r").read().strip())
data_day8 = [re.findall(r'([a-z]{3}) ([+|-]\d+)',x)[0] for x in data_day8]

class BootCode:
    def __init__(self,start,ins):
        self.instructions = ins
        self.pos = start
        self.accumulator = 0
        self.visited = [0]
    
    def second_visit(self):
        return self.pos in self.visited
    
    def jump(self,steps):
        self.pos += int(steps)
    
    def acc(self,amount):
        self.pos += 1
        self.accumulator += int(amount)
        
    def operate(self,pair):
        if pair[0] == 'acc':
            self.acc(pair[1])
        elif pair[0] == 'jmp':
            self.jump(pair[1])
        elif pair[0] == 'nop':
            self.pos +=1
    
    def run(self):
        while 1==1:
            self.operate(self.instructions[self.pos])
            if self.is_finished():
                break
            if self.second_visit():
                break
            else:
                self.visited +=[self.pos]
        return self.is_finished(),self.accumulator
    
    def is_finished(self):
        return self.pos == len(self.instructions)
    
    def get_visited(self):
        return self.visited

A = BootCode(0,data_day8)
print(f'Part 1: Accumulator = {A.run()[1]}')

# Part 2
flip_set = [i for i in A.get_visited() if data_day8[i][0] in ['nop','jmp']]

def do_flip(d,pos):
    if d[pos][0] == 'jmp':
        d[pos] = ('nop',d[pos][1])
    else:
        d[pos] = ('jmp',d[pos][1])
    return d
    

for j in flip_set:
    d1 = [l for l in data_day8]
    new_ins = do_flip(d1,j)
    B = BootCode(0,new_ins).run()
    if B[0]:
        break
print(f'Part 2: Accumulator = {B[1]}')
```

```python
# Day 12
import re,math
input_12 = open("inputs/day12.txt", "r").read()
small_12 = """F10
N3
F7
R90
F11"""
def get_day12_data(data):
    data = re.split('\n',data.strip())
    data = [re.findall(r'([A-Z])(\d+)',x)[0] for x in data]
    data = [(x[0],int(x[1])) for x in data]
    return data

class Ship:
    def __init__(self,instructions,facing='E',waypoint=None):
        self.compass = 'ESWN'
        self.x = 0
        self.y = 0
        self.dir = self.compass.find(facing)
        self.instructions = instructions
        self.wpmode = False
        if waypoint:
            (self.wpx,self.wpy) = waypoint
            self.wpmode = True 
        
    def move(self,instruction):
        if instruction[0] == 'N':
            self.y += instruction[1]
        elif instruction[0] == 'S':
            self.y -= instruction[1]
        elif instruction[0] == 'E':
            self.x += instruction[1]
        elif instruction[0] == 'W':
            self.x -= instruction[1]
        elif instruction[0] == 'L':
            turn_90 = int(instruction[1]/90)
            self.dir = (self.dir-turn_90)%4
        elif instruction[0] == 'R':
            turn_90 = int(instruction[1]/90)
            self.dir = (self.dir+turn_90)%4
        elif instruction[0] == 'F':
            self.move((self.compass[self.dir],instruction[1]))
    
    def move_with_waypoint(self,instruction):
        if instruction[0] == 'N':
            self.wpy += instruction[1]
        elif instruction[0] == 'S':
            self.wpy -= instruction[1]
        elif instruction[0] == 'E':
            self.wpx += instruction[1]
        elif instruction[0] == 'W':
            self.wpx -= instruction[1]
        
        elif instruction[0] in ['L','R']:
            teta = (instruction[1]) * (-1 if instruction[0]=='L' else 1) * 2 * (math.pi/360)
            new_x = math.sin(teta)*self.wpy + math.cos(teta)*self.wpx
            new_y = math.cos(teta)*self.wpy - math.sin(teta)*self.wpx
            (self.wpx,self.wpy) = (round(new_x),round(new_y))
        
        elif instruction[0] == 'F':
            self.x += instruction[1]*self.wpx
            self.y += instruction[1]*self.wpy
            
    def sail(self):
        for command in self.instructions:
            if self.wpmode:
                self.move_with_waypoint(command)
            else:
                self.move(command)
        return abs(self.x)+abs(self.y)

S = Ship(get_day12_data(input_12)).sail()
print(f'Part 1: {S}')

# Part 2
S_WP = Ship(get_day12_data(input_12),'E',(10,1)).sail()
print(f'Part 1: {S_WP}')
```
